```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

```python
# Load dataset (update path if using a CSV)
df = pd.read_csv("dataset.csv")  # Or use the DataFrame directly if already in memory
```

```python
df.isnull().sum()
```

```python
# Drop non-feature columns
df_model = df.drop(columns=["campaign_id", "start_date", "end_date"])

# One-hot encode categorical columns
df_encoded = pd.get_dummies(df_model, drop_first=True)
```

```python
# Features and Target
X = df_encoded.drop(columns=["estimated_revenue"])
y = df_encoded["estimated_revenue"]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=8)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Predict and evaluate
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print("Model Performance:")
print(f"R² Score: {r2:.4f}")

# Save the model
joblib.dump(model, "estimated_revenue_model.pkl")
print("Model saved as 'estimated_revenue_model.pkl'")
```
